#**Tensors Math**

```python
import torch
```

```python
X = torch.tensor([1,2,3])
Y = torch.tensor([9,8,7])
```

```python
# Addition

z1 = torch.empty(3)
torch.add(X,Y,out=z1)
print(z1)
```

```python
hh = torch.add(X, Y)
print(hh)
```

```python
z = X + Y
print(z)
```

```python
# Subtraction

Z = X - Y
print(Z)
```

```python
# Division

z = torch.true_divide(X, Y)
print(z)
```

```python
# Inplace Operations

t = torch.zeros(3)
t.add_(X)
print(t)
```

```python
t += X
print(t)
```

```python
# Exponentiation

q = X.pow(2)
print(q)
```

```python
q = X ** 2
print(q)
```

```python
# Simple Comparison

X = torch.tensor([1,2,3])
z = X > 0
print(z)
```

```python
z = X < 0
print(z)
```

```python
# Matrix Multiplication

x1 = torch.rand((2,5))
x2 = torch.rand((5,3))
```

```python
x3 = torch.mm(x1, x2)   #2x3
print(x3)
```

```python
x3 = x1.mm(x2)
print(x3)
```

```python
#Matrix Exponentiation

matrix_exp = torch.rand(5, 5)
print(matrix_exp.matrix_power(3))
```

```python
# Element wise Multiplication

z = X * Y
print(z)
```

```python
# Dot product

z = torch.dot(X, Y)
print(z)
```

```python
# Batch Matrix Multiplication

batch = 32
n = 10
m = 20
p = 30

tensor1 = torch.rand((batch, n, m))
tensor2 = torch.rand((batch, m, p))
out_bmm = torch.bmm(tensor1, tensor2)  #(batch, n, p)
print(out_bmm)
```

```python
#Example of Broadcasting

x1 = torch.rand((5,5))
x2 = torch.rand((1,5))

z = x1 - x2
print(z)
z = x1 ** x2
print(z)
```

```python
# Oher useful tensor operations
sum_x = torch.sum(X, dim=0)
print(sum_x)

values, indices = torch.max(X, dim=0)
print(values, indices)

values, indices = torch.min(X, dim=0)
print(values, indices)

abs_x = torch.abs(x)
print(abs_x)

z = torch.argmax(X, dim=0)
print(z)

z = torch.argmin(X, dim=0)
print(z)

mean_x = torch.mean(X.float(), dim=0)
print(mean_x)

z = torch.eq(X, Y)
print(z)
```

```python
z = torch.sort(Y, dim=0, descending=False)
print(z)
```

```python
z = torch.sort(Y, dim=0, descending=True)
print(z)
```

```python
sorted_y, indices = torch.sort(Y, dim=0, descending=False)
print(sorted_y, indices)
```

```python
z = torch.clamp(X, min=0)
print(z)
```

```python
z = torch.clamp(X, min=0, max=10)
print(z)
```

```python
x = torch.tensor([1,0,1,1,1], dtype=bool)
z = torch.any(x)
print(z)
z = torch.all(x)  # False because we have one value zero
print(z)
```

#**Tensor Indexing**

```python
batch_size = 10
features= 25
x = torch.rand((batch_size, features))
print(x)
```

```python
print(x[0])
```

```python
print(x[0].shape)  #x[0,:]
```

```python
print(x[:,0])
print(x[:,0].shape)
```

```python
print(x[2, 0:10])
print(x[2, 0:10].shape)
```

```python
x[0, 0] = 100
```

```python
# Fancy Indexing
x = torch.arange(10)
indices = [2,5,8]
print(x[indices])
```

```python
x = torch.rand(3,5)
rows = torch.tensor([1,0])
cols = torch.tensor([4,0])
print(x[rows, cols])
print(x[rows, cols].shape)
```

```python
# More advanced indexing
x = torch.arange(10)
print(x[(x < 2) | (x > 8)])
print(x[(x < 2) & (x > 8)])
print(x[(x < 2) ^ (x > 8)])
```

```python
print(x[x.remainder(2) == 0])
```

```python
# Other useful operations
print(torch.where(x > 5, x, x*2))
```

```python
print(torch.tensor([0,0,1,1,1,2,2,2,2]).unique())
```

```python
print(x.ndimension())  #5x5x5
print(x.numel())
```

#**Tensors Reshaping**

```python
x = torch.arange(9)
```

```python
x_3x3 = x.view(3, 3)
print(x_3x3)
print(x_3x3.shape)
```

```python
x_3x3 = x.reshape(3, 3)
print(x_3x3)
```

```python
y = x_3x3.t()
print(y)
print(y.contiguous().view(9))
```

```python
x1 = torch.rand(2, 5)
x2 = torch.rand(2, 5)
print(torch.cat((x1, x2), dim=0).shape)
print(torch.cat((x1, x2), dim=1).shape)
```

```python
z = x1.view(-1)
print(z)
print(z.shape)
```
